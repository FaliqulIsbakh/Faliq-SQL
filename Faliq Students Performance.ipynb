```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("StudentsPerformance.csv")

df.head()
```

Hasil Interpretasi:

Dataset StudentsPerformance berisi data 1000 siswa dengan informasi gender, kelompok etnis, tingkat pendidikan orang tua, jenis lunch, status persiapan ujian, serta nilai matematika, membaca, dan menulis. Tampilan df.head() menunjukkan 5 data pertama sebagai sampel dari keseluruhan dataset.

```
print("Jumlah Baris dan Kolom:")
print(df.shape)

print("\nNama Kolom:")
print(df.columns)

print("\nTipe Data:")
print(df.dtypes)

print("\nInformasi Dataset:")
print(df.info())
```

Hasil Interpretasi:

Berdasarkan hasil pemeriksaan dataset, terdapat 1000 data siswa dengan 8 kolom. Lima kolom berisi data kategori yaitu gender, race/ethnicity, parental level of education, lunch, dan test preparation course, sedangkan tiga kolom lainnya berisi data numerik yaitu math score, reading score, dan writing score. Seluruh kolom memiliki 1000 data lengkap tanpa nilai kosong (missing value), sehingga dataset sudah siap digunakan untuk analisis.

```
print(df.isnull().sum() ) #missing value
```

Hasil Interpretasi:

Berdasarkan hasil pengecekan missing value, semua kolom memiliki nilai 0, artinya tidak terdapat data yang kosong pada dataset. Kolom gender, race/ethnicity, parental level of education, lunch, test preparation course, math score, reading score, dan writing score semuanya memiliki data lengkap sebanyak 1000 baris. Maka, dataset tidak memerlukan proses pembersihan data terkait nilai kosong.





```
print(df['gender'].value_counts())

sns.countplot(data=df, x='gender')
plt.title('Jumlah Siswa Berdasarkan Gender')
plt.show()
```

Hasil Interpretasi:

Berdasarkan hasil analisis jumlah siswa berdasarkan gender, terdapat 518 siswa perempuan (female) dan 482 siswa laki-laki (male). Hal ini menunjukkan bahwa jumlah siswa perempuan sedikit lebih banyak dibandingkan siswa laki-laki, dengan selisih 36 siswa. Secara keseluruhan, distribusi gender dalam dataset cukup seimbang.

```
print(df['lunch'].value_counts())

sns.countplot(data=df, x='lunch')
plt.title('Jumlah Siswa Berdasarkan Jenis Lunch')
plt.show()
```

Hasil Interpretasi:

Berdasarkan hasil analisis jumlah siswa berdasarkan jenis lunch, terdapat 645 siswa dengan kategori standard dan 355 siswa dengan kategori free/reduced. Hal ini menunjukkan bahwa sebagian besar siswa berada pada kategori standard lunch, sedangkan siswa dengan kategori free/reduced jumlahnya lebih sedikit. Distribusi jenis lunch pada dataset menunjukkan adanya perbedaan jumlah yang cukup terlihat antara kedua kategori.


```
persentase = df['gender'].value_counts(normalize=True)*100
print(persentase)

persentase.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Persentase Gender')
plt.show()
```

Hasil Interpretasi:

Berdasarkan hasil analisis persentase gender, dari total siswa terdapat 51,8% siswa perempuan (female) dan 48,2% siswa laki-laki (male). Persentase ini menunjukkan bahwa jumlah siswa perempuan sedikit lebih tinggi dibandingkan siswa laki-laki, namun perbedaannya tidak terlalu besar sehingga distribusi gender dalam dataset tergolong cukup seimbang.

```
sns.scatterplot(
    data=df,
    x='math score',
    y='reading score'
)

plt.title('Math Score vs Reading Score')
plt.show()
```

Hasil Interpretasi:

Berdasarkan grafik Math Score vs Reading Score, terlihat bahwa terdapat hubungan positif antara nilai matematika dan nilai membaca. Artinya, siswa yang memperoleh nilai matematika tinggi cenderung juga memiliki nilai membaca yang tinggi. Sebagian besar data terkumpul pada rentang nilai sekitar 50–80, sehingga dapat disimpulkan bahwa kemampuan matematika dan membaca memiliki pola yang saling berkaitan

```
sns.scatterplot(
    data=df,
    x='math score',
    y='writing score',
    hue='gender'
)

plt.title('Math Score vs Writing Score Berdasarkan Gender')
plt.show()
```

Hasil Interpretasi:

Berdasarkan scatter plot, terdapat hubungan positif antara math score dan writing score, di mana semakin tinggi nilai matematika, semakin tinggi pula nilai menulis. Selain itu, siswa perempuan (female) cenderung memiliki nilai menulis yang lebih tinggi dibandingkan siswa laki-laki (male) pada rentang nilai matematika yang sama. Hal ini menunjukkan adanya perbedaan performa menulis berdasarkan gender.

```
sns.scatterplot(
    data=df,
    x='reading score',
    y='writing score',
    hue='test preparation course'
)

plt.show()
```

Hasil Interpretasi:

Berdasarkan scatter plot, terdapat hubungan positif yang kuat antara reading score dan writing score, di mana semakin tinggi nilai membaca, semakin tinggi pula nilai menulis. Selain itu, siswa yang menyelesaikan test preparation course (completed) cenderung memiliki nilai membaca dan menulis yang lebih tinggi dibandingkan siswa yang tidak mengikuti atau tidak menyelesaikan kursus (none).

```
rata_gender = df.groupby('gender')[
    ['math score','reading score','writing score']
].mean()

print(rata_gender)

rata_gender.plot(kind='bar')
plt.title('Rata-rata Nilai Berdasarkan Gender')
plt.show()
```

Hasil Interpretasi:

Berdasarkan grafik rata-rata nilai berdasarkan gender, siswa laki-laki memiliki rata-rata math score yang lebih tinggi (68,73) dibandingkan siswa perempuan (63,63). Sebaliknya, siswa perempuan memiliki rata-rata reading score (72,61) dan writing score (72,47) yang lebih tinggi dibandingkan siswa laki-laki (65,47 dan 63,31). Hal ini menunjukkan bahwa laki-laki cenderung lebih unggul dalam matematika, sedangkan perempuan cenderung lebih unggul dalam membaca dan menulis.

```
rata_test = df.groupby('test preparation course')[
    ['math score','reading score','writing score']
].mean()

print(rata_test)

rata_test.plot(kind='bar')
plt.title('Rata-rata Nilai Berdasarkan Test Preparation')
plt.show()
```

Hasil Interpretasi:

Berdasarkan grafik rata-rata nilai berdasarkan test preparation course, siswa yang menyelesaikan kursus persiapan (completed) memiliki rata-rata nilai matematika (69,70), membaca (73,89), dan menulis (74,42) yang lebih tinggi dibandingkan siswa yang tidak mengikuti atau tidak menyelesaikan kursus (none). Hal ini menunjukkan bahwa mengikuti kursus persiapan tes berpengaruh positif terhadap peningkatan hasil belajar siswa.

```
sns.violinplot(
    data=df,
    x='gender',
    y='math score'
)

plt.show()
```

Hasil Interpretasi:

Berdasarkan grafik violin plot, distribusi math score siswa laki-laki dan perempuan memiliki pola yang relatif mirip. Namun, median nilai matematika siswa laki-laki terlihat sedikit lebih tinggi dibandingkan siswa perempuan. Selain itu, sebagian besar nilai matematika kedua kelompok berada pada rentang 60–80, yang menunjukkan bahwa mayoritas siswa memperoleh nilai pada rentang tersebut.

```
sns.violinplot(
    data=df,
    x='lunch',
    y='reading score'
)
```

Hasil Interpretasi:

Berdasarkan grafik violin plot, siswa dengan kategori makan siang standard memiliki median reading score yang lebih tinggi dibandingkan siswa dengan kategori free/reduced. Selain itu, distribusi nilai membaca pada kelompok standard cenderung berada pada rentang nilai yang lebih tinggi. Hal ini menunjukkan bahwa siswa yang mendapatkan makan siang standar cenderung memiliki kemampuan membaca yang lebih baik dibandingkan siswa yang menerima makan siang gratis atau bersubsidi.

```
corr = df[['math score',
           'reading score',
           'writing score']].corr()

plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Korelasi Antar Nilai')
plt.show()
```

Hasil Interpretasi:

Berdasarkan heatmap korelasi, ketiga jenis nilai memiliki hubungan positif yang kuat. Korelasi tertinggi terdapat antara reading score dan writing score sebesar 0,95, yang menunjukkan bahwa siswa dengan nilai membaca tinggi cenderung memiliki nilai menulis yang tinggi. Sementara itu, math score juga berkorelasi kuat dengan reading score (0,82) dan writing score (0,80). Hal ini menunjukkan bahwa peningkatan pada satu kemampuan akademik cenderung diikuti oleh peningkatan pada kemampuan akademik lainnya.
